# Day 1

```python

f = open('aoc_2023_d1_input1.txt')
#f = open('aoc_2023_d1_test_input.txt')
ln = f.readline()
n = 0
ans = 0
while ln != "":
  n += 1
  if n >= 10000: #for testing/debug, increase to more than actual file size for real
    break
  l = [ln[a] for a in range(0,len(ln)-1)] ## remove newline
  #print(l)
  ll = len(l)
  #print(ll)
  this_num_chrs = ""
  i = 0 #number of chars processed in this string
  for c in l:
    #print(i, ' = ', c)
    if c.isdigit():
      this_num_chrs += c
    elif ll > i+2:
      #read next 2 chars and check for one, two and six
      wrd = l[i] + l[i+1] + l[i+2]
      #print(wrd)
      if wrd == "one":
        this_num_chrs += '1'
      elif wrd == "two":
        this_num_chrs += '2'
      elif wrd == "six":
        this_num_chrs += '6'
      #if not, read one more char and check for four, five and nine
      elif ll > i+3:
        wrd += l[i+3]
        #print(wrd)
        if wrd == "four":
          this_num_chrs += '4'
        elif wrd == "five":
          this_num_chrs += '5'
        elif wrd == "nine":
          this_num_chrs += '9'
        #if not, read one more char and check for three, seven and eight
        elif ll > i+4:
          wrd += l[i+4]
          #print(wrd)
          if wrd == "three":
            this_num_chrs += '3'
          elif wrd == "seven":
            this_num_chrs += '7'
          elif wrd == "eight":
            this_num_chrs += '8'
    i += 1
  #print(this_num_chrs)
  x = this_num_chrs[0] + this_num_chrs[len(this_num_chrs) - 1]
  #print(x)
  ans += int(x)
  ln = f.readline()
print(ans)
f.close
```

```python
import sys

# -----------------------------------------------------------------------
# getNextToken()
# Returns: token position, token
# -----------------------------------------------------------------------
def getNextToken(str, delims, offset):
  # print('getNextToken str: ', str)
  # print('getNextToken delims: ', delims)
  # print('getNextToken offset: ', offset)
  l = len(str)
  # print('getNextToken str len: ', l)
  if l == 0:
    return (-1, '')
  else:
    # We're comparing the offset against the length of the string
    # Offset only goes up to len(str)-1
    # Set l to match
    l -= 1
  if offset >= l:
    return (-2, 'Offset too large')
  i = offset
  # print('getNextToken str[i]: ', str[i])
  # If offset is on a delimiter, start from the next non-delim character
  if str[i] in delims:
    while str[i] in delims:
      i += 1
  tkn_pos = i  # Token Position: where the token was found
  tkn = ''
  while i <= l:
    # Add character to token if not a newline or delim
    if str[i] == '\n':
      #print('Found New Line at position ', i) # Print EOL position
      break
    elif str[i] in delims:
      break
    else:
      tkn += str[i]
    i += 1
  # print('getNextToken i: ', i)
  # print('getNextToken tkn: ', tkn)
  # print('getNextToken tkn_pos: ', tkn_pos)
  return (tkn_pos, tkn)
# -----------------------------------------------------------------------
# getNextToken
# -----------------------------------------------------------------------

# -----------------------------------------------------------------------
# tokenize()
# Returns: list of (token position, token) tuples
# -----------------------------------------------------------------------
# Tokenize whole string if no offset provided
def tokenize(str, delims, offset = 0):
  tokens = []
  l = len(str)
  if l == 0:
    return []    # Exception, check for empty list in calling function
  else:
    # We're comparing the offset against the length of the string
    # Offset only goes up to len(str)-1
    # Set l to match
    l -= 1
  if offset >= l:
    return []    # Exception, check for empty list in calling function
  ln_offset = offset
  x = 0                     # Debug
  while ln_offset < l:      # Note: ln_offset is incremented by getNextToken return value
    # x += 1                  # Debug: limit number of tokens
    # if x > 10:               # Debug: limit number of tokens
    #   sys.quit()            # Debug: limit number of tokens
    # print('tokenize: ln_offset', ln_offset)
    token = getNextToken(str, delims, ln_offset)
    if token[1] == 'Offset too large':
      print('Unexpected end of string')
      break
    tokens.append(token)
    ln_offset = token[0] + len(token[1])  # Calculate next offset
    # sys.quit()
    #print('offset', ln_offset, 'len l', len(str))
  return tokens
# -----------------------------------------------------------------------
# tokenize
# -----------------------------------------------------------------------

# Unit Test tokenize
# ------------------
# my_list = tokenize('', '')
# print(my_list)
# if len(my_list) == 0:
#   print('Empty list returned')
# sys.quit()

# -----------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------
f = open('aoc_2023_d2_input1.txt')
#f = open('aoc_2023_d2_test1.txt')
ln = f.readline()
delims = [' ', ',', ';']
gn = 0                     # For number of lines in file (equals game no.)
posb_games = 0
tot_pwr = 0
while ln != '':
  gn += 1
  if gn >= 1000:          # Debug: limit number of lines
    print('Debug Stop')   # Debug: limit number of lines
    sys.quit()            # Debug: limit number of lines

  # Removing newline is not really required when using getNextToken()
  #l = [ln[a] for a in range(0,len(ln)-1)] ## remove newline

  # print(ln)

  # Start after colon
  ln_offset = ln.find(':') + 1
  if ln_offset == -1:
    print('No starting offset')
    break
  l = len(ln)-1
  ln_tokens = tokenize(ln, delims, ln_offset)
  hi_red = 0
  hi_grn = 0
  hi_blu = 0
  num_cubes = 0
  pwr_cubes = 0
  for i in range(0, len(ln_tokens)):
    if ln_tokens[i][1].isdigit():
      num_cubes = int(ln_tokens[i][1])
    else:
      if ln_tokens[i][1] == 'red' and num_cubes > hi_red:
        hi_red = num_cubes
      elif ln_tokens[i][1] == 'green' and num_cubes > hi_grn:
        hi_grn = num_cubes
      elif ln_tokens[i][1] == 'blue' and num_cubes > hi_blu:
        hi_blu = num_cubes
  pwr_cubes = hi_red * hi_grn * hi_blu
  tot_pwr += pwr_cubes
  if hi_red <= 12 and hi_grn <= 13 and hi_blu <= 14:
    posb_games += gn
  # print('Main Stop')
  # sys.quit()
  ln = f.readline()
print('Possible Games: ', posb_games)
print('Total Power: ', tot_pwr)
f.close()
# -----------------------------------------------------------------------
# END
# -----------------------------------------------------------------------
```
