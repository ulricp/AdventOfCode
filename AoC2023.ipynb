# Day 1

```python

f = open('aoc_2023_d1_input1.txt')
#f = open('aoc_2023_d1_test_input.txt')
ln = f.readline()
n = 0
ans = 0
while ln != "":
  n += 1
  if n >= 10000: #for testing/debug, increase to more than actual file size for real
    break
  l = [ln[a] for a in range(0,len(ln)-1)] ## remove newline
  #print(l)
  ll = len(l)
  #print(ll)
  this_num_chrs = ""
  i = 0 #number of chars processed in this string
  for c in l:
    #print(i, ' = ', c)
    if c.isdigit():
      this_num_chrs += c
    elif ll > i+2:
      #read next 2 chars and check for one, two and six
      wrd = l[i] + l[i+1] + l[i+2]
      #print(wrd)
      if wrd == "one":
        this_num_chrs += '1'
      elif wrd == "two":
        this_num_chrs += '2'
      elif wrd == "six":
        this_num_chrs += '6'
      #if not, read one more char and check for four, five and nine
      elif ll > i+3:
        wrd += l[i+3]
        #print(wrd)
        if wrd == "four":
          this_num_chrs += '4'
        elif wrd == "five":
          this_num_chrs += '5'
        elif wrd == "nine":
          this_num_chrs += '9'
        #if not, read one more char and check for three, seven and eight
        elif ll > i+4:
          wrd += l[i+4]
          #print(wrd)
          if wrd == "three":
            this_num_chrs += '3'
          elif wrd == "seven":
            this_num_chrs += '7'
          elif wrd == "eight":
            this_num_chrs += '8'
    i += 1
  #print(this_num_chrs)
  x = this_num_chrs[0] + this_num_chrs[len(this_num_chrs) - 1]
  #print(x)
  ans += int(x)
  ln = f.readline()
print(ans)
f.close
```
