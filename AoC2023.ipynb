# Day 1

```python

f = open('aoc_2023_d1_input1.txt')
#f = open('aoc_2023_d1_test_input.txt')
ln = f.readline()
n = 0
ans = 0
while ln != "":
  n += 1
  if n >= 10000: #for testing/debug, increase to more than actual file size for real
    break
  l = [ln[a] for a in range(0,len(ln)-1)] ## remove newline
  #print(l)
  ll = len(l)
  #print(ll)
  this_num_chrs = ""
  i = 0 #number of chars processed in this string
  for c in l:
    #print(i, ' = ', c)
    if c.isdigit():
      this_num_chrs += c
    elif ll > i+2:
      #read next 2 chars and check for one, two and six
      wrd = l[i] + l[i+1] + l[i+2]
      #print(wrd)
      if wrd == "one":
        this_num_chrs += '1'
      elif wrd == "two":
        this_num_chrs += '2'
      elif wrd == "six":
        this_num_chrs += '6'
      #if not, read one more char and check for four, five and nine
      elif ll > i+3:
        wrd += l[i+3]
        #print(wrd)
        if wrd == "four":
          this_num_chrs += '4'
        elif wrd == "five":
          this_num_chrs += '5'
        elif wrd == "nine":
          this_num_chrs += '9'
        #if not, read one more char and check for three, seven and eight
        elif ll > i+4:
          wrd += l[i+4]
          #print(wrd)
          if wrd == "three":
            this_num_chrs += '3'
          elif wrd == "seven":
            this_num_chrs += '7'
          elif wrd == "eight":
            this_num_chrs += '8'
    i += 1
  #print(this_num_chrs)
  x = this_num_chrs[0] + this_num_chrs[len(this_num_chrs) - 1]
  #print(x)
  ans += int(x)
  ln = f.readline()
print(ans)
f.close
```

```python
f = open('aoc_2023_d2_input1.txt')

def getNextToken(str, delims, offset):
  # print('getNextToken str: ', str)
  # print('getNextToken delims: ', delims)
  # print('getNextToken offset: ', offset)
  l = len(str)
  # print('getNextToken str len: ', l)
  if l == 0:
    return (0, '')
  if offset >= l:
    return (0, 'Offset too large')
  i = offset
  # print('getNextToken str[i]: ', str[i])
  # If offset is on a delimiter, start from the next character
  if str[i] in delims:
    i += 1
  tkn = ''
  while i < l:
    # Add character to token if not a delimiter or newline
    if str[i] not in delims and str[i] != '\n':
      token += str[i]
    i += 1
    # print('getNextToken i: ', i)
    print('getNextToken tkn: ', tkn)
  return (i, tkn)

ln = f.readline()
delims = [' ', ',', ';']
n = 0                     # number of lines in file
while ln != '':
  n += 1
  if n >= 3:          #for testing/debug, increase to more than actual file size for real
    print('Debug Stop')
    break

  # Removing newline is not really required when using getNextToken()
  #l = [ln[a] for a in range(0,len(ln)-1)] ## remove newline

  print(ln)

  # Start after colon
  ln_offset = ln.find(':') + 1
  if ln_offset == -1:
    print('No starting offset')
    break
  tokens = []
  while ln_offset < len(ln):       # Note: ln_offset is incremented by getNextToken return value
    token = getNextToken(ln, delims, ln_offset)
    if token[1] == 'Offset is too large':
      print('Unexpected end of string')
      break
    tokens.append(token[1])
    ln_offset = token[0]
  for i in tokens:
    print(tokens.index(i), i)
  # print(tokens)
  ln = f.readline()
f.close()
```
